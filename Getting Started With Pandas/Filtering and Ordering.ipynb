### Filtering and Ordering Columns and Rows

```python
import pandas as pd
```

```python
df = pd.read_json("data.json")
df.head()
```

```python
## Filter based on Columns
df[df['api_seq']=='AccessRestriction']
```

```python
specific_entries = ['AccessRestriction','AbstractDemo']

df[df['api_seq'].isin(specific_entries)]
```

```python
## The contains

df[df['api_seq'].str.contains('AccessRestriction', na=False)]
```

## Set Index

```python
df2 = df.set_index('id')
```

```python
df2
```

```python
### Filter

df2.filter(items=['summary','api_seq'])
```

```python
### Filter

df2.filter(items=['f6f12135-52f4-47be-aabd-c3743355788e'], axis= 0 )
```

```python
### Like & Filter
df2.filter(like = 'f6f12135-52f4-47be-aabd-c3743355788e', axis= 0 )
```

```python
#Loc location
df2.loc['f6f12135-52f4-47be-aabd-c3743355788e']
```

```python
#iLoc 
df2.iloc[3]
```

```python
#Order by
df[df['id']].sort_values(by=['id'], ascending=False)
```
