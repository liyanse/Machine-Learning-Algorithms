## Clustering

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import recall_score
```

```python
#For the object data type columns, get dummies
df1 = pd.get_dummies(data, drop_first = True, columns=['Education','Marital_Status'])

df1
```

```python
#drop the 'Z_CostContact', 'Z_Revenue','Dt_Customer', which with no meaning
df2 = df1.drop(['Z_CostContact', 'Z_Revenue','Dt_Customer'], axis=1)
```

```python
df2
```

```python
X = df2[
    ['Income', 'Kidhome', 'Teenhome', 'Recency',
       'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts',
       'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases',
       'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases',
       'NumWebVisitsMonth', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5',
       'AcceptedCmp1', 'AcceptedCmp2', 'Complain', 
       'Education_Basic', 'Education_Graduation', 'Education_Master',
       'Education_PhD', 'Marital_Status_Alone', 'Marital_Status_Divorced',
       'Marital_Status_Married', 'Marital_Status_Single',
       'Marital_Status_Together', 'Marital_Status_Widow',
       'Marital_Status_YOLO']]

y = df2['Response']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1998)
```

```python
X_train.info()
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
def ACC(predict, model_name):
    accuracy = np.mean(predict == y_test)
    print("Accuracy of", model_name, "is", round(float(accuracy), 3))
    
def TPR(predict,model_name):
    TPR = recall_score(y_test, predict)
    print("TPR of", model_name, "is", round(float(TPR), 3))
```

```python
def knn():
    knn_classifier = KNeighborsClassifier(n_neighbors = 7)
    knn_classifier.fit(X_train, y_train)
    pred_Y = knn_classifier.predict(X_test)
    return ACC(pred_Y, 'knn'), TPR(pred_Y, 'knn')
```

```python
print(knn())
```

## Plotting Clusters

```python
import scipy.cluster.hierarchy as sch

from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering 
from sklearn.neighbors import LocalOutlierFactor
from sklearn import preprocessing
```

```python
data['TotalProduct']=(data['MntWines']+data['MntFruits']+data['MntMeatProducts']+data['MntFishProducts']+data['MntSweetProducts']+data['MntGoldProds'])
data['TotalPurchases']=(data['NumDealsPurchases']+data['NumWebPurchases']+data['NumCatalogPurchases']+data['NumStorePurchases'])
```

```python
df_cluster=data[['Income','TotalProduct','TotalPurchases','Recency']].reset_index(drop =True)
```

```python
df_cluster
```

```python
plt.figure(figsize=[20,8])

plt.scatter(df_cluster.iloc[:,1] , df_cluster.iloc[:,0] , s=100 , color = 'blue')
plt.title('The scatterplot of Customers Income and number of buying products')
plt.xlabel('TotalBuyProduct')
plt.ylabel('Income')
plt.show()
```

```python
clf = LocalOutlierFactor(n_neighbors=2)
df_cluster['Outlier']=clf.fit_predict(df_cluster)
df_cluster=df_cluster[df_cluster['Outlier']==1]
df_cluster.drop(['Outlier'], axis=1, inplace = True)
df_cluster.reset_index(drop =True,inplace=True)
```

```python
df_cluster.describe()
```

```python
plt.figure(figsize=[20,8])

plt.scatter(df_cluster.iloc[:,1] , df_cluster.iloc[:,0] , s=100 , color = 'blue')
plt.title('The scatterplot of Customers Income and number of buying products')
plt.xlabel('TotalBuyProduct')
plt.ylabel('Income')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=0).fit(df_cluster[['TotalProduct','TotalPurchases','Recency']])
kmeans.labels_
```

```python
df_cluster.head()
```

```python
#filter rows of original data
filtered_label0 = df_cluster[kmeans.labels_==0]
filtered_label1 = df_cluster[kmeans.labels_==1]
filtered_label2 = df_cluster[kmeans.labels_==2]
filtered_label3 = df_cluster[kmeans.labels_==3]
filtered_label4 = df_cluster[kmeans.labels_==4]
```

```python
plt.figure(figsize=[20,8])

plt.subplot(1,3,1)
plt.scatter(filtered_label0.iloc[:,1] , filtered_label0.iloc[:,0] , s=100 , color = 'red')
plt.scatter(filtered_label1.iloc[:,1] , filtered_label1.iloc[:,0] , s=100 , color = 'blue')
plt.scatter(filtered_label2.iloc[:,1] , filtered_label2.iloc[:,0] , s=100 , color = 'green')
plt.scatter(filtered_label3.iloc[:,1] , filtered_label3.iloc[:,0] , s=100 , color = 'cyan')
plt.scatter(filtered_label4.iloc[:,1] , filtered_label4.iloc[:,0] , s=100 , color = 'yellow')
plt.title('Clusters of Customers (Clustering Kmean Model)')
plt.xlabel('TotalBuyProduct')
plt.ylabel('Income')
```

```python
df_cluster['Kgroup']=kmeans.labels_
```

```python
for i in ("Income","TotalProduct","TotalPurchases","Recency"):
    plt.figure(figsize=[20,8])
    plt.subplot(1,3,1)
    sns.boxplot(x="Kgroup", y=i, data=df_cluster, palette="Set1")
```
